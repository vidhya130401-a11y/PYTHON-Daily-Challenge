## Student Profile Creator


```python
# Create variable
student_name = input("Enter your name: ")
student_age = input("Enter your age: ")
favourate_subject = ("Math", "Science", "English")
```

```python
# Display student information
print("\n--- student profile ---")
print(f"Name: {student_name}")
print(f"Age: {student_age}")
print(f"favourate_subject: {', '. join(favourate_subject) }")
```

```python
# ASk for a new favourite subject
new_subject = input("\nEnter a new favourite subject to add: ")
favourate_subject = list(favourate_subject)
favourate_subject.append(new_subject)
favourate_subject = tuple(favourate_subject) # Convert back to tuple if desired, or keep as list
```

```python
# Display student list
print("\n--- Updated Student Profile ---")
print(f"Name: {student_name}")
print(f"Age: {student_age}")
print(f"favorte subject: {', '.join(favourate_subject)}")
```
