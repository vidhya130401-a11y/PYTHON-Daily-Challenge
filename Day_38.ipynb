## Day - 38 (representing feedback)

```python
# user input
feedback = input("Enter your feedback: ")
```

```python
# cound the number of words
count = len(feedback.split())
print(f"Total words are: {count}")
```

```python
# count "data" appers in the sentence, reardless of case.
data_count = feedback.lower().count("data")
print(f"count of the word 'data' is: {data_count}")
```

```python
# Extract and print the first 5 and last 5 characters
first_five = feedback[:5]
last_five = feedback[-5]
print(f"first 5 characters are: {first_five}")
print(f"last 5 characters are: {last_five}")
```

```python
# Reverse the sentence
reversed_sentence = feedback[::-1]
print(reversed_sentence)
```
