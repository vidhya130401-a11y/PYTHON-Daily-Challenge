## Tuple,Immutale vs mutable, list creation


```python
# Create a tuple with fixed product catrgories
fixed_products = ("Electrronics", "Stationary", "Clothing")
print(fixed_products)
```

```python
# Create a list of products that can be updated dynamically
products = ["Laptop", "Mobile Phone", "Headphones"]
print(products)
```

```python
# Ask the user to enter a new product and add it to the list
new_product = input("Enter a new product: ")
products.append(new_product)
print(products)
```

```python
# Ask the user to input the index of a product to remove it from the list
try:
    index_to_remove = int(input("Enter the index of the product to remove (e.g., 0 for Laptop, 1 for Mobile Phone): "))
    if 0 <= index_to_remove < len(products):
        removed_product = products.pop(index_to_remove)
        print(f"Removed '{removed_product}' from the list.")
    else:
        print("Invalid index. Please enter a valid index within the current range.")
except ValueError:
    print("Invalid input. Please enter a number for the index.")

print("Updated products list:", products)
```

```python
# final
print("final product categories (imputable):", fixed_products)
print("final product (mutable)", products)
```
