```python
contacts = [
    ["vidhya", "9629333995", "vidhya@001"],
    ["shakar", "1234567890", "shakar@002"],
    ["tamil", "8494643929", "tamil@123"],
]
```

```python
def add_contact():
  name = input("Enter name: ")
  phone = input("Enter phone: ")
  email = input("Enter email: ") # Changed 'contacts' to 'email'
  contacts.append([name, phone, email])
  print("contacts added successfully!\n")
```

```python
def display_sorted_contacts():
  sorted_contacts = sorted(contacts, key=lambda x: x[0])
  print("\nall contacts (sorted): ")
  for c in sorted_contacts:
    print(f"name: {c[0]}, phone: {c[1]}, email: {c[2]}")
    print()
```
