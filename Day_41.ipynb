## Student Grades & Hobbies Tracker

```python
# Create dictionary for student grades values
student_grades = {
    "vidhya": 85,
    "shakar": 90,
    "tamil": 78
  }
print(student_grades)
```

```python
# Add or update grades for at least 3 students
student_grades["vidhya"] = 85
student_grades['shakar'] = 90

#Add a new student
student_grades["siva"] = 98
print(student_grades)
```

```python
# create a set of hobbies
hobbies = set()
# add at least 3 hobbies
hobbies.add("reading")
hobbies.add("cycling")
hobbies.add("painting")
print("hobbies set: ", hobbies)
```

```python
# Convert hobbies set into frozenset
frozen_hobbies = frozenset(hobbies)
print("frozen hobbies: ", frozen_hobbies)
```
