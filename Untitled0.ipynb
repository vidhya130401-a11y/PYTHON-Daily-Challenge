## BMI CALCULATOR

```python
# INPUT
weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))
```

```python
# validate input
if weight <= 0 or height <= 0:
   print("weight and height must be positive numbuer. ")
```

```python
# calculate bmi
bmi = weight / (height ** 2)
```

```python
from IPython.core.compilerop import CachingCompiler
# classify bmi
if bmi <18.5:
    category = "underweight"
elif bmi < 24.9:
    category = "normal weghit"
elif bmi < 29.9:
     category = "overweghit"
else:
    category = "obese"
```

```python
# display result
print(f" your bmi {bmi:.2f}. you are calssified as: {category}")
```
